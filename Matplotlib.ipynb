```python
# Create some data -Stateless way of plotting # pyplot of of plotting
x = [1,2,3,4]
y = [11,22,33,44]
```

```python
# With a semi colon and now a y value
plt.plot(x,y);
```

```python
#  Creating a plot with the OO version, confusing way first
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#2
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# Easier and more robust going forward (what we're going to use)

fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig,ax = plt.subplots(figsize = (10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Sample Simple plot", xlabel = "x-axis", ylabel = "y-axis")

#5. Save and show
fig.savefig("Images/simple-plot.png")
```
