```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig,ax = plt.subplots(figsize = (10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Sample Simple plot", xlabel = "x-axis", ylabel = "y-axis")

#5. Save and show
fig.savefig("Images/simple-plot.png")
```

```python
import numpy as np
```

### Line 
Line is the default type of visualization in Matplotlib


```python
# Create an array
x = np.linspace(0,10,100)
x[:10]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
#plot the data
fig,ax = plt.subplots()
ax.plot(x,x*x)
```

## Scatter 

```python
# Need to recreate our figure and axis instances when we want a new figure
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

## Bar

-Vertical
-Horizontal

```python
#plots can be made from dictionary
nut_butter_prices = {"Almond butter": 10, "Peanut butter": 8, "Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "My butter store", ylabel= "Price($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title = "My butter store", ylabel= "Price($)")
```
